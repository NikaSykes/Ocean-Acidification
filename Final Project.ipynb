# East Coast

```python
#Import data file, skiprows and usecols will vary based on specific data file

f=open ('pHMiamiFL.csv', 'r')
data = np.loadtxt('pHMiamiFL.csv', comments="END_DATA", delimiter=",",skiprows=32, usecols=(29))
f.close()

#Remove -999 outliers

newdataX=[]
for x in range (0,len(data)):
    if ((data[x])>0):
                newdataX.append(data[x])

#Cacluate the average pH level             

avepH= (sum(newdataX))/ float(len(newdataX))

#Print average and range

print('pH average:',(avepH))
print('pH range:', ([min(newdataX),max(newdataX)]))

#Plot the pH levels on the y, with data reading along the x axis

Y = newdataX
plt.plot(Y)
plt.xlabel('Data Reading')
plt.ylabel('pH ')
```

# West Coast
